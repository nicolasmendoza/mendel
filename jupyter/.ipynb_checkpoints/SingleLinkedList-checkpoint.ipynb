```python
"""
Linked list accommodate items of varying sizes and allow easy insertion and deletion of items.

linux kernel linked list data structure implementation in C avalaible in: 
https://github.com/torvalds/linux/blob/master/include/linux/list.h

"""

# Single LinkedList, each item points to its successor.

class Element():
    """Represent a element in a linkedList dataStructure"""
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class LinkedList(object):
    """LinkedList data structre"""
    
    def __init__(self, head=None):
        self.list_head = head
        
    def append(self, new_element):
        """Append a element to the linkedList"""
        head = self.list_head
        
        if self.list_head:
            while head.next:
                head = head.next
            head.next = new_element
        
    def insert(self, new_element, position):
        """Insert a new node at the given position."""
        counter = 1
        current = self.list_head
        if position > 1:
            while current and counter < position:
                if counter == position - 1:
                    new_element.next = current.next
                    current.next = new_element
                current = current.next
                counter += 1
        elif position == 1:
            new_element.next = self.list_head
            self.list_head = new_element

    def delete(self, value):
        current = self.list_head
        previous = None
        while current.value != value and current.next:
            previous = current
            current = current.next
        if current.value == value:
            if previous:
                previous.next = current.next
            else:
                self.list_head = current.next       
        
    def get_position(self, position):
        """Returns element of x position"""
        if self.list_head:
            head = self.list_head
            for i in range(position):
                current_head = head.next
            return head
        else:
            return None      
```

```python
e1 = Element(1)
e2 = Element(2)
e3 = Element(3)
e4 = Element(4)

ll = LinkedList(e1)
ll.append(e2)
ll.append(e3)

print(ll.list_head.next.next.value)
print(ll.get_position(3).value)

ll.insert(e4,3)
print(ll.get_position(3).value)

ll.delete(1)
print(ll.get_position(1).value)
print(ll.get_position(2).value)
print(ll.get_position(3).value)
```
